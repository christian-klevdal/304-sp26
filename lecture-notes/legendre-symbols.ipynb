# Quadratic Residues, Legendre Symbols, and Gauss's Lemma

This lecture introduces quadratic residues modulo primes, the Legendre symbol as a tool for determining when a number is a square, and Gauss's Lemma as a method for computing the Legendre symbol.

### Quadratic Residues

<div class="theorem" style="border: 1px solid #ccc; padding: 10px; margin: 5px 0; background-color: #f9f9f9; border-radius: 5px; overflow: hidden;">
    <p style="font-size: 1.2em; font-weight: bold; margin-top: 10px;">Definition (Quadratic Residue)</p>
    <p>Let $n$ be a positive integer and $a$ be an integer. We say that $a$ is a <em>quadratic residue modulo $n$</em> if there exists an integer $x$ such that 
    $$x^2 \equiv a \pmod{n}.$$
    If no such $x$ exists, then $a$ is called a <em>quadratic non-residue modulo $n$</em>.</p>
</div>

In other words, $a$ is a quadratic residue modulo $n$ if $a$ is a "square" modulo $n$.

**Examples:**

- _Modulo 7:_ The squares modulo 7 are:
  - $1^2 \equiv 1 \pmod{7}$
  - $2^2 \equiv 4 \pmod{7}$
  - $3^2 \equiv 9 \equiv 2 \pmod{7}$
  - $4^2 \equiv 16 \equiv 2 \pmod{7}$
  - $5^2 \equiv 25 \equiv 4 \pmod{7}$
  - $6^2 \equiv 36 \equiv 1 \pmod{7}$
  
  So the quadratic residues modulo 7 are $\{1, 2, 4\}$, and the non-residues are $\{3, 5, 6\}$.

- _Modulo 11:_ We have $1^2 = 1$, $2^2 = 4$, $3^2 = 9$, $4^2 = 5$, $5^2 = 3$ (all mod 11). So the quadratic residues modulo 11 are $\{1, 3, 4, 5, 9\}$.

Let's verify these computationally:

```python
# Verify quadratic residues modulo 7
p = 7
print(f"Quadratic residues modulo {p}:")
residues = set()
for x in range(1, p):
    residue = (x^2) % p
    residues.add(residue)
    print(f"{x}^2 ≡ {residue} (mod {p})")

print(f"\nQuadratic residues mod {p}: {sorted(residues)}")
non_residues = set(range(1, p)) - residues
print(f"Quadratic non-residues mod {p}: {sorted(non_residues)}")
```

Here's a function to compute all quadratic residues modulo any integer $n$:

```python
def quadratic_residues(n):
    """
    Return the set of quadratic residues modulo n.
    """
    residues = set()
    for x in range(n):
        residues.add((x^2) % n)
    return sorted(residues)

# Test with various moduli
for n in [5, 11, 13]:
    qr = quadratic_residues(n)
    print(f"Quadratic residues mod {n}: {qr}")
    print(f"Number of QR mod {n}: {len(qr)}\n")
```

### The Legendre Symbol

To work efficiently with quadratic residues modulo a prime $p$, we introduce the Legendre symbol.

<div class="theorem" style="border: 1px solid #ccc; padding: 10px; margin: 5px 0; background-color: #f9f9f9; border-radius: 5px; overflow: hidden;">
    <p style="font-size: 1.2em; font-weight: bold; margin-top: 10px;">Definition (Legendre Symbol)</p>
    <p>Let $p$ be an odd prime and $a$ be an integer. The <em>Legendre symbol</em> $\left( \frac{a}{p} \right)$ is defined as:
    $$\left( \frac{a}{p} \right) = \begin{cases}
    1 & \text{if } a \text{ is a quadratic residue modulo } p \text{ and } a \not\equiv 0 \pmod{p}, \\
    -1 & \text{if } a \text{ is a quadratic non-residue modulo } p, \\
    0 & \text{if } a \equiv 0 \pmod{p}.
    \end{cases}$$
    </p>
</div>

_Important Note:_ The Legendre symbol $\left( \frac{a}{p} \right)$ looks like a fraction, but it is _not_ a fraction! It's simply notation for indicating whether $a$ is a square modulo $p$.

**Examples:**
- $\left( \frac{1}{7} \right) = 1$ because $1 \equiv 1^2 \pmod{7}$
- $\left( \frac{2}{7} \right) = 1$ because $2 \equiv 3^2 \pmod{7}$
- $\left( \frac{3}{7} \right) = -1$ because $3$ is not a quadratic residue modulo $7$
- $\left( \frac{4}{7} \right) = 1$ because $4 \equiv 2^2 \pmod{7}$

#### Properties of the Legendre Symbol

The Legendre symbol has several important properties that make it useful for computations:

<div class="theorem" style="border: 1px solid #ccc; padding: 10px; margin: 5px 0; background-color: #f9f9f9; border-radius: 5px; overflow: hidden;">
    <p style="font-size: 1.2em; font-weight: bold; margin-top: 10px;">Theorem (Multiplicativity)</p>
    <p>Let $p$ be an odd prime and $a, b$ be integers. Then
    $$\left( \frac{ab}{p} \right) = \left( \frac{a}{p} \right) \left( \frac{b}{p} \right).$$
    </p>
</div>

This means that the Legendre symbol is _multiplicative_. For example:
- $\left( \frac{6}{7} \right) = \left( \frac{2 \cdot 3}{7} \right) = \left( \frac{2}{7} \right) \left( \frac{3}{7} \right) = (1)(-1) = -1$

Let's compute some Legendre symbols:

```python
def legendre_symbol(a, p):
    """
    Compute the Legendre symbol (a/p) by checking if a is a QR mod p.
    """
    a = a % p
    if a == 0:
        return 0
    
    # Check if a is a quadratic residue
    for x in range(1, p):
        if (x^2) % p == a:
            return 1
    return -1

# Compute Legendre symbols for various values
p = 11
print(f"Legendre symbols modulo {p}:")
for a in range(1, p):
    symbol = legendre_symbol(a, p)
    print(f"({a}/{p}) = {symbol:2d}")

print(f"\nVerify multiplicativity: (6/11) = (2/11) * (3/11)")
print(f"(6/11) = {legendre_symbol(6, 11)}")
print(f"(2/11) * (3/11) = {legendre_symbol(2, 11)} * {legendre_symbol(3, 11)} = {legendre_symbol(2, 11) * legendre_symbol(3, 11)}")
```

### Euler's Criterion

While the definition of the Legendre symbol is straightforward, checking all squares to determine if $a$ is a quadratic residue is inefficient. Euler's Criterion provides a much faster method.

<div class="theorem" style="border: 1px solid #ccc; padding: 10px; margin: 5px 0; background-color: #f9f9f9; border-radius: 5px; overflow: hidden;">
    <p style="font-size: 1.2em; font-weight: bold; margin-top: 10px;">Theorem (Euler's Criterion)</p>
    <p>Let $p$ be an odd prime and $a$ be an integer with $\gcd(a, p) = 1$. Then
    $$\left( \frac{a}{p} \right) \equiv a^{\frac{p-1}{2}} \pmod{p}.$$
    </p>
</div>

This gives us a computational method: to determine if $a$ is a quadratic residue modulo $p$, simply compute $a^{(p-1)/2} \bmod p$:
- If the result is $1$, then $a$ is a quadratic residue
- If the result is $-1 \equiv p-1 \pmod{p}$, then $a$ is a non-residue

_Why does this work?_ By Fermat's Little Theorem, $a^{p-1} \equiv 1 \pmod{p}$, so $a^{(p-1)/2}$ is a square root of $1$ modulo $p$. The only square roots of $1$ modulo $p$ are $\pm 1$.

Let's verify Euler's Criterion:

```python
# Verify Euler's Criterion
def legendre_euler(a, p):
    """
    Compute Legendre symbol using Euler's Criterion.
    """
    result = power_mod(a, (p-1)//2, p)
    if result == p - 1:
        return -1
    else:
        return result

# Test Euler's Criterion against our previous method
p = 13
print(f"Verifying Euler's Criterion for p = {p}:\n")
print("a  | (a/p) by definition | a^((p-1)/2) mod p | Match?")
print("-" * 60)
for a in range(1, p):
    symbol_def = legendre_symbol(a, p)
    symbol_euler = legendre_euler(a, p)
    match = "✓" if symbol_def == symbol_euler else "✗"
    print(f"{a:2d} | {symbol_def:20d} | {symbol_euler:17d} | {match}")

print(f"\n✓ Euler's Criterion verified for all a from 1 to {p-1}!")
```

### Gauss's Lemma

Gauss's Lemma provides another method for computing the Legendre symbol that's based on counting negative residues in a specific set.

#### The Set $S(a,p)$

<div class="theorem" style="border: 1px solid #ccc; padding: 10px; margin: 5px 0; background-color: #f9f9f9; border-radius: 5px; overflow: hidden;">
    <p style="font-size: 1.2em; font-weight: bold; margin-top: 10px;">Definition</p>
    <p>Let $p$ be an odd prime and $a$ be an integer with $\gcd(a, p) = 1$. Define
    $$S(a, p) = \left\{a \bmod_{\pm} p, \, 2a \bmod_{\pm} p, \, 3a \bmod_{\pm} p, \, \ldots, \, \frac{p-1}{2}a \bmod_{\pm} p\right\}$$
    where $x \bmod_{\pm} p$ denotes the unique integer in the range $\left[-\frac{p}{2}, \frac{p}{2}\right]$ that is congruent to $x$ modulo $p$.</p>
</div>

In other words, we take the multiples $a, 2a, 3a, \ldots, \frac{p-1}{2}a$ and reduce them modulo $p$ to lie in the symmetric interval around zero.

**Example:** Let $p = 7$ and $a = 3$. Then:
- $1 \cdot 3 = 3 \equiv 3 \pmod{7}$ → in range $[-3, 3]$: $3$
- $2 \cdot 3 = 6 \equiv 6 \equiv -1 \pmod{7}$ → in range $[-3, 3]$: $-1$
- $3 \cdot 3 = 9 \equiv 2 \pmod{7}$ → in range $[-3, 3]$: $2$

So $S(3, 7) = \{3, -1, 2\}$.

Let's write code to compute $S(a, p)$:

```python
def compute_S(a, p):
    """
    Compute the set S(a, p) as defined in Gauss's Lemma.
    Returns elements reduced to the range [-(p-1)/2, (p-1)/2].
    """
    S = []
    for k in range(1, (p+1)//2):
        # Compute k*a mod p
        val = (k * a) % p
        # Reduce to symmetric range
        if val > p // 2:
            val = val - p
        S.append(val)
    return S

# Example: compute S(3, 7)
a, p = 3, 7
S = compute_S(a, p)
print(f"S({a}, {p}) = {S}")

# Count negative elements
mu = sum(1 for x in S if x < 0)
print(f"μ({a}, {p}) = {mu} (number of negative elements)")

# Try a few more examples
print("\nMore examples:")
for a, p in [(2, 7), (5, 11), (3, 13)]:
    S = compute_S(a, p)
    mu = sum(1 for x in S if x < 0)
    print(f"S({a}, {p}) = {S}, μ = {mu}")
```

#### Statement of Gauss's Lemma

The pattern you may have noticed in the exploration above is formalized by Gauss's Lemma:

<div class="theorem" style="border: 1px solid #ccc; padding: 10px; margin: 5px 0; background-color: #f9f9f9; border-radius: 5px; overflow: hidden;">
    <p style="font-size: 1.2em; font-weight: bold; margin-top: 10px;">Theorem (Gauss's Lemma)</p>
    <p>Let $p$ be an odd prime and $a$ be an integer with $\gcd(a, p) = 1$. Let $\mu(a, p)$ be the number of negative elements in $S(a, p)$. Then
    $$\left( \frac{a}{p} \right) = (-1)^{\mu(a,p)}.$$
    </p>
</div>

In other words:
- If $\mu(a,p)$ is _even_, then $\left( \frac{a}{p} \right) = 1$ (so $a$ is a quadratic residue)
- If $\mu(a,p)$ is _odd_, then $\left( \frac{a}{p} \right) = -1$ (so $a$ is a non-residue)

**Example:** We computed $S(3, 7) = \{3, -1, 2\}$, which has $\mu(3, 7) = 1$ negative element. Thus:
$$\left( \frac{3}{7} \right) = (-1)^1 = -1$$
So $3$ is a quadratic non-residue modulo $7$.

Let's verify Gauss's Lemma for several values:

```python
# Verify Gauss's Lemma
def legendre_gauss(a, p):
    """
    Compute Legendre symbol using Gauss's Lemma.
    """
    S = compute_S(a, p)
    mu = sum(1 for x in S if x < 0)
    return (-1)^mu

# Verify Gauss's Lemma against Euler's Criterion
p = 17
print(f"Verifying Gauss's Lemma for p = {p}:\n")
print("a  | μ(a,p) | (-1)^μ | (a/p) Euler | Match?")
print("-" * 55)
for a in range(1, min(p, 12)):  # Test first few values
    S = compute_S(a, p)
    mu = sum(1 for x in S if x < 0)
    gauss = legendre_gauss(a, p)
    euler = legendre_euler(a, p)
    match = "✓" if gauss == euler else "✗"
    print(f"{a:2d} | {mu:6d} | {gauss:6d} | {euler:11d} | {match}")

print(f"\n✓ Gauss's Lemma verified!")
```
