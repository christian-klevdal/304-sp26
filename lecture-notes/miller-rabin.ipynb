# Miller-Rabin Primality Test

This notebook implements the Miller-Rabin primality test and demonstrates how to check if a number is a witness for compositeness.

```python
def miller_rabin_witness(n, a):
    """
    Determine if 'a' is a Miller-Rabin witness for the compositeness of n.
    
    Returns:
        True if a is a witness (n is definitely composite)
        False if a is not a witness (n passes the test for base a)
    
    The Miller-Rabin test:
    Write n - 1 = 2^s * d where d is odd
    n passes the test if either:
      - a^d ≡ 1 (mod n), or
      - a^(2^r * d) ≡ -1 (mod n) for some 0 ≤ r < s
    """
    if n < 2:
        return True  # Not prime
    
    if n == 2:
        return False  # 2 is prime
    
    if n % 2 == 0:
        return True  # Even numbers > 2 are composite
    
    # Write n - 1 as 2^s * d
    s = 0
    d = n - 1
    while d % 2 == 0:
        s += 1
        d //= 2
    
    # Compute a^d mod n
    x = pow(a, d, n)
    
    # Check if a^d ≡ 1 (mod n)
    if x == 1:
        return False  # n passes the test
    
    # Check if a^(2^r * d) ≡ -1 (mod n) for some r
    for r in range(s):
        if x == n - 1:  # x ≡ -1 (mod n)
            return False  # n passes the test
        x = pow(x, 2, n)
    
    # If neither condition is met, a is a witness
    return True
```

```python
def is_miller_rabin_probable_prime(n, a):
    """
    Check if n is a probable prime for base a.
    Returns True if n passes the Miller-Rabin test (probably prime).
    Returns False if a is a witness (n is definitely composite).
    """
    return not miller_rabin_witness(n, a)
```

## Examples

Let's test various numbers with different bases to see if they are witnesses.

```python
# Example 1: Test n = 25 (composite) with base a = 7
n = 25
a = 7

print(f"Testing n = {n} with base a = {a}")
is_witness = miller_rabin_witness(n, a)
print(f"Is {a} a witness for {n}? {is_witness}")
print(f"Does {n} pass the test for base {a}? {not is_witness}")
print()
```

```python
# Example 2: Test n = 45 (composite) with base a = 2
n = 45
a = 2

print(f"Testing n = {n} with base a = {a}")
is_witness = miller_rabin_witness(n, a)
print(f"Is {a} a witness for {n}? {is_witness}")
print(f"Does {n} pass the test for base {a}? {not is_witness}")
print()
```

```python
# Example 3: Test n = 97 (prime) with base a = 5
n = 97
a = 5

print(f"Testing n = {n} (a prime) with base a = {a}")
is_witness = miller_rabin_witness(n, a)
print(f"Is {a} a witness for {n}? {is_witness}")
print(f"Does {n} pass the test for base {a}? {not is_witness}")
print()
```

```python
# Example 4: Test Carmichael number 561 with various bases
n = 561
print(f"Testing Carmichael number n = {n}")
print(f"(Note: 561 = 3 × 11 × 17 is composite but passes Fermat test for all bases)\n")

for a in [2, 3, 5, 7, 11]:
    is_witness = miller_rabin_witness(n, a)
    print(f"Base a = {a}: Is witness? {is_witness}, Passes test? {not is_witness}")
```

## Testing All Bases for a Given Number

Let's see which bases are witnesses for a composite number.

```python
# Find all witnesses for n = 91 (composite, 7 × 13)
n = 49
witnesses = []
non_witnesses = []

for a in range(1, n):
    if miller_rabin_witness(n, a):
        witnesses.append(a)
    else:
        non_witnesses.append(a)

print(f"Testing n = {n}")
print(f"\nNumber of witnesses: {len(witnesses)}")
print(f"Number of non-witnesses: {len(non_witnesses)}")
print(f"\nFirst 10 witnesses: {witnesses[:10]}")
print(f"First 10 non-witnesses: {non_witnesses[:10]}")
```

```python
# Compare with a prime number
n = 89  # prime
witnesses = []

for a in range(2, min(n, 50)):
    if miller_rabin_witness(n, a):
        witnesses.append(a)

print(f"Testing n = {n} (prime)")
print(f"\nNumber of witnesses found (should be 0 for primes): {len(witnesses)}")
if witnesses:
    print(f"Witnesses: {witnesses}")
else:
    print("No witnesses found - this is expected for a prime number!")
```
